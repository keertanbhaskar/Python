```python
# filtering

import pandas as pd

data = {
  "Name":["ram","sham","yami","aditi",'suresh'],
  "Age":[20,45,36,37,26],
  "Salary":[50000,43000,67000,70000,58000]
}

df = pd.DataFrame(data)

high_salary = df[df['Salary'] > 50000]
print('Employee with salary > 50000')
print(high_salary)

# multicondition with and 
print(15*"-","age > 30 and salary > 50000", 15*'-')
filtered = df[(df["Age"]>30) & (df["Salary"] > 50000)]
print(filtered)

# multicondition with or
print(15*"-","age > 30 or salary > 50000", 15*'-')
filtered1 = df[(df["Age"]>30) | (df["Salary"] > 50000)]
print(filtered1)
```
