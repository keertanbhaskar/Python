```python
# Handling missing data
# NaN(Not a number), None(for object data types)
# isnull() ,True -> NaN is missing

import pandas as pd

workers_Info = {
    
    "Name": ["Amit", 'shyam', "Rahul", "Sneha", "Karan", "Anjali", "Vikram", "Meera"],
    "Age": [25, 30, 28, 35, 40, None, 32, 29],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing", "HR", "Finance", "IT"],
    "Salary": [35000, None, 50000, 75000, 65000, 40000, 55000, 62000],
    "Experience": [2, 5, 4, 8, 10, 3, 6, 5],
    "City": ["Chennai", "Bangalore", "Hyderabad", "Chennai", "Mumbai", "Pune", "Delhi", "Bangalore"]
}

df = pd.DataFrame(workers_Info)
display(df.isnull())

print(25 * "=", "total of missing values", 25 * "=")

display(df.isnull().sum())
```

```python
# filling default values in the missing places => data cleaning

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'] = df['Age'].astype(int)


df['Salary'] = df['Salary'].fillna(df['Salary'].mean()).astype(int)
display(df)
```

```python
# interpolation => fill estimated value  => for numerical values
# linear,polynomial, time
# while working with time series data
# numeric data with trend
# avoid dropping rows
# dis => it won't work with categorical data

import pandas as pd

df = pd.DataFrame({
    "Sales": [100, None, 300]
})


df["Sales"] = df["Sales"].interpolate()
print(df)
```

### sorting & aggregation

```python
# sorting data
# df.sort_values(by="column name",true, inplace=True) => true for ascending

import pandas as pd

employees = {
    "Employee_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Amit", "Rahul", "Sneha", "Karan", "Anjali",
             "Vikram", "Meera", "Priya", "Arjun", "Neha"],
    "Age": [25, 30, 30, None, 40, 32, 29, 25, None, 31],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing",
                   "HR", "Finance", "IT", "Sales", "Marketing"],
    "Salary": [35000, 60000, 50000, 75000, None,
               40000, 55000, 62000, 48000, 70000],
    "Experience": [2, 5, 4, 8, 10, 3, 6, 5, 4, 7],
    "City": ["Chennai", "Bangalore", "Hyderabad", "Chennai", "Mumbai",
             "Pune", "Delhi", "Bangalore", "Chennai", "Mumbai"]
}

df = pd.DataFrame(employees)

df.sort_values(by='Age',ascending=True,inplace=True)
display(df)
```

```python
# multiple condition sorting

df.sort_values(by=['Age','Salary'],ascending=[True,False],inplace=True)
print('sorted by age ascending  and salary by descending ')
display(df)
```

```python
# summary statistics
# df['Column Name'].mean(),df['Column Name'].max(),df['Column Name'].min()

avg_sal = df['Salary'].mean()
print(avg_sal)
```

```python
# grouping
grouped = df.groupby('Age')['Salary'].sum()
print(grouped)

# for multiple columns
# grouped = df.groupby(['Age','Name'])['Salary'].sum
```

```python
# aggregation => sum(),mean(),count(),min(),max()
# merging & concatenating => pd.merge(df1,df2,on='column name', how='type of join')

import pandas as pd
df_customer = pd.DataFrame({
  'CusID':[1,2,3],
  'Name':['Sam','Ram','Shyam']
})

df_orders = pd.DataFrame({
  'CusID':[1,2,4],
  'OrderAmount':[240,450,350]
})


# merger
df_merged = pd.merge(df_customer,df_orders,on='CusID',how='inner') #outer,left,right,cross
print(df_merged)

```
