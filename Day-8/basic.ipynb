### numpy array and basics


```python
import numpy as np
```

```python
marks = np.array([85, 90, 78, 92, 60])
print("marks:",marks)

arr_2d = np.array([[1,1,2],[3,5,6]])
print("2d-Array:",arr_2d)
```

### list vs numpy array

```python
py_list = [1,2,3,4]
print("python list multiplication",py_list*2)

print(20*"=")

np_array = np.array([1,2,3]) #element wise multiplication
print("numpy array mul:",np_array*2)
```

#### Filtering and slicing

```python
passed = marks[marks >= 70]
print("Passed Students:",passed)
print("First 3 Marks:",marks[:3])
print("Last Mark:",marks[-1])
```

#### Built-In math functions


```python
print("Total:",np.sum(marks))
print("Average:",np.mean(marks))
print("Highest:",np.max(marks))
print("Lowest:",np.min(marks))
```

#### 2D Array

```python
sub_marks = np.array([
  [85,90], #student 1
  [78,88], #st 2
  [92,95] #st 3
])
print("Marks Table:\n",sub_marks)
```

### Row-wise and column-wise operations

```python
# total marks per student => axis1 - row wise, axis 0=> column wise
print("Total per student:",np.sum(sub_marks,axis=1))

print("Average per subject:",np.mean(sub_marks,axis=0))
```

### more numpy utilities

```python
print(np.arange(1,10))
print(np.zeros((2,3)))
print(np.ones((3,2)))
print(np.sort(marks))
```

### NumPy Score Analyzer

```python
Marks = np.array([75,85,60,45])

# print avg,total,max,min
print("Average:",np.mean(Marks))
print("Total:",np.sum(Marks))
print("Max:",np.max(Marks))
print("Min:",np.min(Marks))

# print student who scored above 60
print("Score above 60:",Marks[Marks > 60])

# 2D array of marks
subject_marks = np.array([
   [75, 80],
    [85, 90],
    [60, 55]
])

# total per student
print("Total per student:", np.sum(subject_marks, axis=1))

# 6. Average per subject
print("Average per subject:", np.mean(subject_marks, axis=0))
```

### Array properties

```python
a = np.array([1, 2, 3, 4])
print(a.shape )     # dimensions
print(a.ndim )     # number of dimensions
print(a.size )      # total elements
print(a.dtype )     # data type
```
